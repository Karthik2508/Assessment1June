## Different types of anime

```python
#different types of anime
d = dataset['Type'].value_counts()
fig = plt.figure(figsize=(6,6))
plt.pie(d,labels=d.index,autopct='%1.1f%%',explode=(0.1, 0, 0, 0,0,0))
plt.show()
```

## Percentages fo watching,completed,on_hold,dropped for different types of anime 

```python
#the percentages fo watching,completed,on_hold,dropped for different types of anime 
d = dataset.groupby(['Type']).sum()[['Watching','Completed','On-Hold','Dropped']]
plt.figure(figsize=(14,14))
plt.subplot(4, 4, 1)
plt.title(d.index[0])
plt.pie(d.iloc[0],labels=d.columns,labeldistance=1)
plt.subplot(4, 4, 3)
plt.title(d.index[1])
plt.pie(d.iloc[1],labels=d.columns,labeldistance=1)
plt.subplot(4, 4, 5)
plt.title(d.index[2])
plt.pie(d.iloc[2],labels=d.columns,labeldistance=1)
plt.subplot(4, 4, 7)
plt.title(d.index[3])
plt.pie(d.iloc[3],labels=d.columns,labeldistance=1)
plt.subplot(4, 4, 9)
plt.title(d.index[4])
plt.pie(d.iloc[4],labels=d.columns,labeldistance=1)
plt.subplot(4, 4, 11)
plt.title(d.index[5])
plt.pie(d.iloc[5],labels=d.columns,labeldistance=1)
plt.show()
```

## Scatterplot for members who still watching animes and who stoped watching animes 

```python
x=dataset['Members']
y=dataset['Watching']
z=dataset['Dropped']
z1=dataset['On-Hold']
plt.figure(figsize=(8,8))
plt.scatter(x,y,label='Watching')
plt.scatter(z,y,color='Red',label='Dropped')
plt.scatter(z1,y,color='Green',label='On-Hold')
plt.legend(loc=1)
plt.show()
```

```python
#from above graph we can say members who dropped watching animes are less compared to members who still watching animes
#members who on-hold watching animes and who dropped watching animes are almost similar 
print('Members : '+str(dataset['Members'].sum()))
print('Watching: '+str(dataset['Watching'].sum()))
print('Dropped : '+str(dataset['Dropped'].sum()))
print('On-Hold : '+str(dataset['On-Hold'].sum()))
```
